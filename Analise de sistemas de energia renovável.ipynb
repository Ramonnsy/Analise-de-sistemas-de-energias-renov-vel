1. Type_of_Renewable_Energy : Código numérico que representa o tipo de fonte de energia renovável (1: Solar, 2: Eólica, 3: Hidrelétrica, 4: Geotérmica, 5: Biomassa, 6: Maré, 7: Onda).
2. Installed_Capacity_MW : Capacidade instalada em megawatts (MW).
3. Energy_Production_MWh : Produção anual de energia em megawatts-hora (MWh).
4. Energy_Consumption_MWh : Consumo anual de energia em megawatts-hora (MWh).
5. Energy_Storage_Capacity_MWh : Capacidade de armazenamento de energia em megawatts-hora (MWh).
6. Storage_Efficiency_Percentage : Eficiência dos sistemas de armazenamento de energia em porcentagem.
7. Grid_Integration_Level : Código numérico que representa o nível de integração da rede (1: Totalmente Integrado, 2: Parcialmente Integrado, 3: Integração Mínima, 4: Microrrede Isolada).
8. Initial_Investment_USD : Custos de investimento inicial em dólares americanos.
9. Funding_Sources : Código numérico que representa a fonte de financiamento (1: Governo, 2: Privado, 3: Parceria Público-Privada).
10. Financial_Incentives_USD : Incentivos financeiros em dólares americanos.
11. GHG_Emission_Reduction_tCO2e : Redução das emissões de gases de efeito estufa em toneladas de CO2 equivalente (tCO2e).
12. Air_Pollution_Reduction_Index : Índice de redução da poluição do ar.
13. Jobs_Created : Número de empregos criados.

Perguntas criadas pelo ChatGPT

1. Qual tipo de fonte de energia renovável tem a maior média de capacidade instalada em megawatts (MW)?
2. Como a eficiência dos sistemas de armazenamento de energia varia entre diferentes tipos de energia renovável?
3. Como o índice de redução da poluição do ar se relaciona com o tipo de fonte de energia renovável e o nível de integração da rede?
4. Os incentivos financeiros têm impacto significativo na criação de empregos na indústria de energia renovável?
5. Qual é a relação da produção anual de energia com o consumo anual.


```python
#Carregando bibliotecas

import pandas as pd
import numpy as np 
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carregando o Dataframe

df = pd.read_csv(os.getcwd() + "\\dataset\\energia\\energy_dataset_.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
#Traduzindo a coluna "Type_of_Renewable_Energy" em seus respectivos valores

Sistemas_Energia = {
    1: "Solar",
    2: "Eólica",
    3: "Hidrelétrica",
    4: "Geotérmica",
    5: "Biomassa",
    6: "Maré",
    7: "Onda"
}

df["Type_of_Renewable_Energy"] = df["Type_of_Renewable_Energy"].replace(Sistemas_Energia)

df.head()
```

# 1. Qual tipo de fonte de energia renovável tem a maior média de capacidade instalada em megawatts (MW)



```python
#Função para agrupar a coluna "Type_of_Renewable_Energy" e a "Installed_Capacity_MW" para ter a sua media.

df.groupby("Type_of_Renewable_Energy")["Installed_Capacity_MW"].mean().sort_values(ascending=False)
```

Aqui podemos ver que a Energia renovavel que tem a maior média de capacaidade instalada em megawatts (MW) é a Hidrelétrica. 

Megawatt é uma unidade de medida de potência, geralmente usada para descrever a quantidade de energia que está sendo produzida ou consumida.

# 2. Como a eficiência dos sistemas de armazenamento de energia varia entre diferentes tipos de energia renovável

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Type_of_Renewable_Energy", y="Storage_Efficiency_Percentage", data=df, palette='RdBu')
plt.xlabel("Energias renováveis")
plt.ylabel("Eficiência do armazenamento de Energia (%)")
plt.title("Eficiência do armazenamento de energia por tipo de energia renovável ")
plt.xticks(rotation=45)
plt.show()
```

Podemos ver que as porcentagens de eficiência de armazenamento para diferentes tipos de energia são até que consistenes, estando principalmente entre 60% a 90% 

# 3. Como o índice de redução da poluição do ar se relaciona com o tipo de fonte de energia renovável e o nível de integração da rede?

```python
#Traduzindo a coluna "Grid_Integration_Level" em seus respectivos valores

integraçao_Rede = {
    1: "Totalmente Integrado",
    2: "Parcialmente Integrado",
    3: "Integração Mínima",
    4: "Microrrede Isolada",
    
}

df["Grid_Integration_Level"] = df["Grid_Integration_Level"].replace(integraçao_Rede)
```

```python
poluição = df.groupby(["Type_of_Renewable_Energy", "Grid_Integration_Level"])["Air_Pollution_Reduction_Index"].mean().unstack()

ax= poluição.plot(kind="bar", figsize=(10, 6))
ax.grid(axis='y')
ax.set_axisbelow(True)

plt.xlabel("Tipo de fonte de energia renovável")
plt.ylabel("Indice de redução de poluição do ar")
plt.title("Índice de Redução da Poluição do Ar por Tipo de Fonte de Energia e Nível de Integração da Rede")

plt.xticks(rotation=45)
plt.legend(title='Nível de Integração da Rede', bbox_to_anchor=(1, 1))
plt.show()
```

Podemos observar que apenas o sistema de energia renovavel "Solar" independe do seu nivel de integração de rede se mostrou acima de 50% 

# 4. Os incentivos financeiros têm impacto significativo na criação de empregos na indústria de energia renovável?

```python
#Traduzindo a coluna "Funding_Sources" em seus respectivos valores

fonte_financiamento = {
    1: "Governo",
    2: "Privado",
    3: "Parceria Público-Privada",    
}
df["Funding_Sources"] = df["Funding_Sources"].replace(fonte_financiamento)
```

```python
emprego = df.groupby("Funding_Sources")["Jobs_Created"].mean().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Funding_Sources", y="Jobs_Created", data=emprego)
ax.grid(axis='y')
ax.set_axisbelow(True)

plt.xlabel('Fonte de Financiamento')
plt.ylabel('Número Médio de Empregos Criados')
plt.title('Número Médio de Empregos Criados por Fonte de Financiamento')
plt.show()
```

Pode-se observar que não tem um impacto significaivo em relação a empregos gerados a partir de diferentes incentivos financeiros.

# 5. Qual é a relação da produçãao anual de energia e o consumo anual

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Energy_Production_MWh", y="Energy_Consumption_MWh", hue="Type_of_Renewable_Energy", data=df, palette="viridis")
plt.title("Produção anual de energia Vs Consumo anual")
plt.xlabel("Produção de energia (MWh)")
plt.ylabel("Consumo de energia (MWh)")
plt.legend(title="Type of Renewable Energy")
plt.show()
```
